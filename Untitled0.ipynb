```python
import os
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("misrakahmed/vegetable-image-dataset")

print("Path to dataset files:", path)
```

```python
# Fungsi untuk memuat dataset dari path yang diberikan
def load_dataset(path):
    """
    Memuat dataset gambar dari direktori yang ditentukan.
    Setiap subdirektori dianggap sebagai label kelas.
    """
    images = []
    labels = []

    # Iterasi melalui setiap folder kelas (misalnya, 'Bean', 'Carrot', dll.)
    for label in os.listdir(path):
        label_path = os.path.join(path, label)
        if not os.path.isdir(label_path):
            continue

        # Iterasi melalui setiap gambar di dalam folder kelas
        for image_file in os.listdir(label_path):
            image_path = os.path.join(label_path, image_file)

            # Baca gambar dalam format grayscale
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            if image is not None:
                # Ubah ukuran gambar menjadi 128x128 piksel
                image = cv2.resize(image, (128, 128))
                images.append(image)
                labels.append(label)

    return np.array(images), np.array(labels)
```

```python
# Tentukan path ke dataset
# Ganti path ini sesuai dengan lokasi dataset di sistem Anda
BASE_PATH = '/kaggle/input/vegetable-image-dataset/Vegetable Images/'
TRAIN_PATH = os.path.join(BASE_PATH, 'train')
VALIDATION_PATH = os.path.join(BASE_PATH, 'validation')
TEST_PATH = os.path.join(BASE_PATH, 'test')
```

```python
# Memuat data dari masing-masing direktori
print("Memuat data latih...")
X_train, y_train = load_dataset(TRAIN_PATH)
print(f"Data latih berhasil dimuat: {X_train.shape[0]} gambar")

print("\nMemuat data validasi...")
X_val, y_val = load_dataset(VALIDATION_PATH)
print(f"Data validasi berhasil dimuat: {X_val.shape[0]} gambar")

print("\nMemuat data uji...")
X_test, y_test = load_dataset(TEST_PATH)
print(f"Data uji berhasil dimuat: {X_test.shape[0]} gambar")
```

```python
# Normalisasi dan Reshape Data
# Mengubah nilai piksel dari [0, 255] menjadi [0, 1] dan meratakan gambar
X_train_norm = X_train.reshape(X_train.shape[0], -1).astype('float32') / 255.0
X_val_norm = X_val.reshape(X_val.shape[0], -1).astype('float32') / 255.0
X_test_norm = X_test.reshape(X_test.shape[0], -1).astype('float32') / 255.0

print(f"\nBentuk data setelah reshape:\nX_train: {X_train_norm.shape}\nX_val: {X_val_norm.shape}\nX_test: {X_test_norm.shape}")
```

```python
# Encoding Label
# Mengubah label teks menjadi angka (misalnya, 'Bean' -> 0, 'Carrot' -> 1)
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Reduksi Dimensi dengan PCA
# Mengurangi jumlah fitur untuk mempercepat training dan mengurangi noise
# PCA di-fit HANYA pada data latih untuk menghindari data leakage
print("\nMelakukan reduksi dimensi dengan PCA...")
pca = PCA(n_components=200, random_state=42)
pca.fit(X_train_norm)
```

```python
# Terapkan transformasi PCA ke semua set data
X_train_pca = pca.transform(X_train_norm)
X_val_pca = pca.transform(X_val_norm)
X_test_pca = pca.transform(X_test_norm)

print(f"Bentuk data setelah PCA:\nX_train_pca: {X_train_pca.shape}\nX_val_pca: {X_val_pca.shape}\nX_test_pca: {X_test_pca.shape}")
```

```python
# Pelatihan Model Logistic Regression
print("\nMelatih model Logistic Regression...")
model = LogisticRegression(solver='saga', max_iter=500, random_state=42, multi_class='multinomial')
model.fit(X_train_pca, y_train_encoded)
print("Model berhasil dilatih.")
```

```python
# Evaluasi Model pada Data Validasi
print("\n--- Evaluasi pada Data Validasi ---")
y_val_pred = model.predict(X_val_pca)
val_accuracy = accuracy_score(y_val_encoded, y_val_pred)
print(f"Akurasi pada data validasi: {val_accuracy:.4f}")
print("Laporan Klasifikasi (Validasi):")
print(classification_report(y_val_encoded, y_val_pred, target_names=label_encoder.classes_))
```

```python
# Evaluasi Akhir pada Data Uji (Test)
print("\n--- Evaluasi Akhir pada Data Uji ---")
y_test_pred = model.predict(X_test_pca)
test_accuracy = accuracy_score(y_test_encoded, y_test_pred)
print(f"Akurasi pada data uji: {test_accuracy:.4f}")
print("Laporan Klasifikasi (Uji):")
print(classification_report(y_test_encoded, y_test_pred, target_names=label_encoder.classes_))
```

```python
# Visualisasi Confusion Matrix untuk Data Uji
print("\nMembuat Confusion Matrix untuk data uji...")
cm = confusion_matrix(y_test_encoded, y_test_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - Data Uji')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.show()
```

```python
# Contoh Prediksi Gambar Tunggal dari Set Uji
# Ambil satu gambar acak dari set data uji asli
random_idx = np.random.randint(0, len(X_test))
sample_image = X_test[random_idx]
sample_label_actual = y_test[random_idx]
```

```python
# Pra-pemrosesan gambar sampel (sama seperti yang dilakukan pada data latih/uji)
sample_image_norm = sample_image.reshape(1, -1).astype('float32') / 255.0
sample_image_pca = pca.transform(sample_image_norm)
```

```python
# Lakukan prediksi
predicted_label_encoded = model.predict(sample_image_pca)
predicted_label = label_encoder.inverse_transform(predicted_label_encoded)[0]
```

```python
# Tampilkan hasilnya
plt.figure(figsize=(4, 4))
plt.imshow(sample_image, cmap='gray')
plt.title(f"Aktual: {sample_label_actual}\nPrediksi: {predicted_label}")
plt.axis('off')
plt.show()
```
